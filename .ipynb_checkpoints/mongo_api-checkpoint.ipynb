```python
import pymongo

password = os.environ.get("PASSWORD_MONGODB")

client = pymongo.MongoClient('mongodb+srv://jose:'+password+'@cluster0.ci561.mongodb.net/sample_airbnb?retryWrites=true&w=majority')
db = client.sample_airbnb

collection = db.listingsAndReviews
```

```python
results = collection.find({}, {"accommodates":1,"bedrooms":1,"beds":1,"number_of_reviews":1,"bathrooms":1,"guests_included":1,"price":1})
```

```python
query = list(results)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(query)
df = df.astype(str).astype(float)
df["_id"] = df._id.astype(str)

df = df[(df.price<3000) & (df.bedrooms<=3)]
df
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.set_title("Accommodates clasification")

for b in np.sort(df.accommodates.unique()):
    ax.scatter(df.bedrooms[df.accommodates==b],df.beds[df.accommodates==b],df.price[df.accommodates==b],label=b)
    
ax.set_xlabel('bedrooms')
ax.set_ylabel('beds')
ax.set_zlabel('price')

ax.legend(loc='center left', bbox_to_anchor=(1.25, 0.5),title="accommodates")
    
plt.show()
```
